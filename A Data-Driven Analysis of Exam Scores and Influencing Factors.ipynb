```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset: https://www.kaggle.com/datasets/lainguyn123/student-performance-factors?resource=download

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
```

```python
df.head()
```

### 1. Histogram

```python
def histogram(df):
    plt.figure(figsize=(8, 6)); sns.histplot(data=df, x='Exam_Score', kde=True)
    plt.title('Distribution of Exam Scores'); plt.xlabel('Exam Score of the Students'); plt.ylabel('Count')
    plt.show()

    
histogram(df)
```

### 2. Line Plot

```python
def lineplot(df): 
    plt.figure(figsize=(8, 6)); plt.scatter(df['Hours_Studied'], df['Exam_Score'], color='blue');
    plt.title('Relationship between Hours Studied and Exam Score');
    plt.xlabel('Hours Studied'); plt.ylabel('Exam Score');
    plt.show()
    
lineplot(df)
```

### 3. Correlation Heatmap

```python
def correlation(df): 
    plt.figure(figsize=(10, 8)); sns.heatmap(df.corr(), annot=True);
    plt.title('Correlation Heatmap');
    plt.show()
    
correlation(df)
```
